```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.integrate import simps

%matplotlib widget
```

## 1nm AlN interlayer

```python
data1 = pd.read_csv('AlGaNchannelHEMT_1nmAlN_withbackbarrier.txt', delimiter='\t')
print(data1.head())
```

```python
x1 = data1['x'].values
n1 = data1['n'].values
Ec1 = data1['Ec'].values
```

```python
plt.close()
plt.plot(x1, n1/np.max(n1), label='n vs x')
plt.plot(x1, Ec1/np.max(Ec1)*4)
plt.xlabel('x')
plt.ylabel('n')
plt.title('Curve of n vs x')
plt.legend()
plt.xlim(100, 400)
plt.ylim(-0.5, 1.25)
plt.grid(True)
plt.show()
```

```python
# Plot the curve for visualization
plt.close()
plt.plot(x1[337:360], Ec1[337:360], marker='o', label='Ec vs x')
plt.xlabel('x')
plt.ylabel('n')
plt.title('Curve of Ec vs x')
plt.legend()
plt.grid(True)
plt.show()

del_Ec1 = np.max(Ec1[337:360]) - np.min(Ec1[337:360])
print('del_Ec:', del_Ec1)
```

```python
# Plot the curve for visualization
barrier_index1 = data1[data1['x'] == 210].index[0]
print('barrier_index:', barrier_index1)
plt.close()
plt.plot(x1[0:barrier_index1+1], n1[0:barrier_index1+1], marker='o', label='n vs x')
plt.xlabel('x')
plt.ylabel('n')
plt.title('Curve of n vs x')
plt.legend()
plt.grid(True)
plt.show()

# Calculate the area under the curve using Simpson's rule
total_area1 = simps(n1, x1)
barrier_area1 = simps(n1[0:barrier_index1+1], x1[0:barrier_index1+1])
```

```python
print(f'Total area under the curve: {total_area1}\nArea under the barrier: {barrier_area1} using Simpson\'s rule')
print(f'Penetration Probability: {barrier_area1/total_area1} using Simpson\'s rule')

print(f'Penetration Probability: {np.trapz(n1[0:barrier_index1+1], x1[0:barrier_index1+1])/np.trapz(n1, x1)} using Trapezoidal rule')
print(f'Error: {np.abs(np.trapz(n1[0:barrier_index1+1], x1[0:barrier_index1+1])/np.trapz(n1, x1) - barrier_area1/total_area1)/(barrier_area1/total_area1)*100}% using Trapezoidal rule')
```

```python
# Plot the curve for visualization
ch_index1 = data1[data1['x'] == 220].index[0]
print('barrier_index:', ch_index1)
plt.close()
plt.plot(x1[ch_index1:], n1[ch_index1:], marker='o', label='n vs x')
plt.xlabel('x')
plt.ylabel('n')
plt.title('Curve of n vs x')
plt.legend()
plt.grid(True)
plt.show()

# Calculate the area under the curve using Simpson's rule
total_area1 = simps(n1, x1)
barrier_area1 = simps(n1[ch_index1:], x1[ch_index1:])

print(f'Total area under the curve: {total_area1}\nArea under the barrier: {barrier_area1} using Simpson\'s rule')
print(f'Penetration Probability: {barrier_area1/total_area1} using Simpson\'s rule')
```

```python
# Plot the curve for visualization
in_index1 = data1[data1['x'] == 210].index[0]
in_index2 = data1[data1['x'] == 220].index[0]+1
print('barrier_index:', ch_index1)
plt.close()
plt.plot(x1[in_index1:in_index2], n1[in_index1:in_index2], marker='o', label='n vs x')
plt.xlabel('x')
plt.ylabel('n')
plt.title('Curve of n vs x')
plt.legend()
plt.grid(True)
plt.show()

# Calculate the area under the curve using Simpson's rule
total_area1 = simps(n1, x1)
in_area1 = simps(n1[in_index1:in_index2], x1[in_index1:in_index2])

print(f'Total area under the curve: {total_area1}\nArea under the barrier: {in_area1} using Simpson\'s rule')
print(f'Penetration Probability: {in_area1/total_area1} using Simpson\'s rule')
```
